```python
import tensorflow as tf
from tensorflow import keras
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images,train_labels), (test_images,test_labels) = fashion_mnist.load_data()
train_images.shape

# label names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# pre process the images
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# Reshape the images.
train_images = np.expand_dims(train_images, axis=3)
test_images = np.expand_dims(test_images, axis=3)
test_images.shape
```

```python
model = keras.Sequential([
keras.layers.Conv2D(32,3,input_shape=(28,28,1),activation="relu"),
keras.layers.Conv2D(64,3,activation="relu"),
keras.layers.MaxPooling2D(2),
keras.layers.Dropout(0.25),
keras.layers.Flatten(),
keras.layers.Dense(128,name="hidden3",activation="relu"),
keras.layers.Dropout(0.5),
keras.layers.Dense(10,name="output")
])
model.summary()
model.save_weights('model.h5')
```

```python
#sparse Categorical takes y_true as numerical , if its one-hot coded then use categorical CE
model.compile(optimizer="Adadelta",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
#model.load_weights('model.h5')
model.fit(train_images,train_labels,batch_size=128,epochs=12,validation_data=(test_images, test_labels))
```

```python
model.evaluate(test_images,test_labels)
```
